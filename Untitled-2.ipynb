```python
data['gender'] = data['gender'].replace({1: 'Women', 2:'Men'})
```

```python
data['cholesterol'] = data['cholesterol'].replace({1: 'normal', 2:'above normal',3:'well above normal'})
```

```python
data['gluc'] = data['gluc'].replace({1: 'normal', 2:'above normal',3:'well above normal'})
```

```python
data['smoke'] = data['smoke'].replace({0: 'non-smoker', 1:'smoker'})
```

```python
data['alco'] = data['alco'].replace({0: 'non-alcoholic', 1:'alcoholic'})
```

```python
data['active'] = data['active'].replace({0: 'non-active', 1:'active'})
```

```python
data['cardio'] = data['cardio'].replace({0: 'yes', 1:'no'})
```

```python


# Specify the percentage of NaNs for each column
# Example: {'column1': 0.05, 'column2': 0.1, ...}
null_percentages = {
    'ap_hi': 0.05,
    'smoke': 0.01,
    'alco': 0.07
    # Add other columns as needed
}

# Iterate over each column and inject NaNs
for col, pct in null_percentages.items():
    # Calculate the number of NaNs to inject
    total_nulls = int(data.shape[0] * pct)

    # Randomly choose row indices
    rows_to_null = np.random.choice(data.index, total_nulls, replace=False)

    # Inject NaN values
    data.loc[rows_to_null, col] = np.nan
```

```python
data.dropna(inplace = True)
```
